```python
Nota = 4.1
Asistencia = 0.85

#Umbrales

Umbral_Nota = 4.0
Umbral_Asistencia = 0.8

if  (Nota > Umbral_Nota) and (Asistencia >= Umbral_Asistencia):
    Resultado = 'Aprobó y tiene excelente asistencia'
elif Nota > Umbral_Nota:
    Resultado = 'Aprobó sin excelente asistencias'
else:
    Resultado = 'Reprobado'
    
print(f'Ejemplo 3: {Nota}, tiene una asistencia del {Asistencia * 100}% => {Resultado}')
```
