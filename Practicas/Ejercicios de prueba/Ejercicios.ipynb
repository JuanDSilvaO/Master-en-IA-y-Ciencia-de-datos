```python
notas = [2.5, 3.1, 4.8, 3.9, 1.7, 4.2, 3.6, 4.9, 2.8, 3.3, 4.1, 3.7, 2.9, 4.5, 1.8, 3.4, 4.0, 2.6, 3.2, 4.7, 2.3, 3.8, 4.4, 1.9, 3.0, 4.6, 2.7, 4.3, 3.5, 2.4, 4.9, 1.6, 3.9, 2.2, 4.8, 3.1, 2.0, 4.5, 3.6, 2.8]

clasificaciones = []

for nota in notas:
    if nota >= 4.5:
        categoria = 'Sobresaliente'
    elif nota >= 3.8:
        categoria = 'Alto'
    elif nota >= 3.0:
        categoria = 'Medio'
    else:
        categoria = 'Bajo'
    clasificaciones.append((round(nota,1),categoria))

print('clasificaciones:')
for nota,cat in clasificaciones:
    print(f' - {nota}: {cat}')

#El _ se us apara indicar que no te importa ese valor
conteo_desempeño = {'Sobresaliente':0 , 'Alto':0 , 'Medio':0 , 'Bajo':0}
for _, cat in clasificaciones:
    conteo_desempeño[cat] += 1
print('Conteo por desempeño:', conteo_desempeño)

```

```python
valores = [10, None, 15, 50, None, 200, 30]
tope = 150

print('procesando con continue y break:')
for v in valores:
    if v is None:
        continue
    if v > tope:
        print(f'Encontrado valor > tope ({v}), deteniendo...')
        break
    print('ok', v)
```

```python
valores = [10, None, 15, 50, None, 200, 30]
objetivo = 500

print(f'Buscando un valor mayor a {objetivo}...')
for v in valores:
    if v is not None and v > objetivo:
        print('Encontrado', v)
        break
else:
    print('No se encontraron valores por encima del objetivo')
```
